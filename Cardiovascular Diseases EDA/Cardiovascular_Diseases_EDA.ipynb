## ASK Question and Answer it

### 1. What is the distribution of heart disease among the individuals in the dataset?

```python
heart_disease_distribution = df['Heart_Disease'].value_counts()



plt.bar(heart_disease_distribution.index,heart_disease_distribution.values,color=custom[1:3])
plt.title('Distribution of Heart Disease')
plt.show()
```

* Heart Disease are way less than from overall

### How many people in the dataset have a history of smoking? 

```python
smoking_history_counts = df['Smoking_History'].value_counts()

# Counting the number of people with a history of smoking
smokers_count = smoking_history_counts['Yes']

# Counting the number of people without a history of smoking
non_smokers_count = smoking_history_counts['No']

# Plotting the counts
plt.bar(['Smokers', 'Non-Smokers'], [smokers_count, non_smokers_count],color=custom[1:3])
plt.xlabel('Smoking History')
plt.ylabel('Count')
plt.title('Number of People with Smoking History')
plt.show()
```

* 45%  are Smokers only (125000)

### Is there a correlation between BMI and the risk of skin cancer?

```python

# Plotting the relationship between BMI and the risk of skin cancer
sns.scatterplot(data=df,x='BMI',y='Skin_Cancer')
plt.xlabel('BMI')
plt.ylabel('Skin Cancer Risk')
plt.title('BMI vs Skin Cancer Risk (Correlation:')
plt.show()
```

### What is the most common age category in the dataset?

```python

Age_cat=df['Age_Category'].value_counts(sort=True).rename_axis('Age_category').reset_index(name='counts').sort_values('Age_category')


#plot
plt.figure(figsize=(10,5))
sns.barplot(data=Age_cat,x='Age_category',y='counts')
plt.title('Most common Age_category')
plt.show() 
```

* Age_range 18-80+
* 50-75 are most common Age in the given data

#### Are there any correlations between height, weight, and the risk of arthritis?

#### Is there any association between fried potato consumption and BMI ?
