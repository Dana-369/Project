## Data Cleaning

```python
# copy the dataset
df=Fifadataset.copy()
```

```python
# Remove the unnecessary newline characters from all columns that have them.
df['Club']=df['Club'].str.replace('\n','')
```

```python
# Strip numbers form the data
def strip_number(series):
    number_only = series.str.extract(r'(\d+)')
    return number_only
```

```python
# Some columns have 'star' characters. Strip those columns of these stars and make the columns numerical
# Convert the height and weight columns to numerical forms

df['Height']=strip_number(df['Height'])
df['Weight']=strip_number(df['Weight'])
df['W/F']=strip_number(df['W/F'])
df['SM']=strip_number(df['SM'])
df['IR']=strip_number(df['IR'])
```

```python
#'Value', 'Wage' and "Release Clause' are string columns. Convert them to numbers. For eg, "M" in value 
#  column is  Million, so multiply the row values by 1,000,000, etc.   




def convert_to_full_number(series):
    # Convert K to 1000 and M to 1000000
    multiplier = {"K": 1000, "M": 1000000}
    result = series.str.extract(r'(\d+\.?\d*)([KM]?)')
    result[0] = result[0].astype(float)
    result[1] = result[1].map(multiplier).fillna(1)
    result = (result[0] * result[1]).astype(int)
    return result
```

```python
df['Value']=convert_to_full_number(df['Value'])
df['Wage']=convert_to_full_number(df['Wage'])
df['Release Clause']=convert_to_full_number(df['Release Clause'])
```

```python
df.info()
```

```python
# Converted to datetime datatype
df['Joined']=pd.to_datetime(df['Joined'],format='%b %d, %Y')
```

```python
# Split Date columns to Day,month, year
df['Joined_date']=df['Joined'].dt.day
df['Joined_month']=df['Joined'].dt.strftime('%b')
df['Joined_year']=df['Joined'].dt.year
df['Joined_monthnumber']=df['Joined'].dt.month
```

```python
# rename columns
df.rename(columns={'Height':'Height(Cm)','Weight':'Weight(Kg)','Value':'Value(Є)','Wage':'Wage(Є)','Release Clause':'Release Clause(Є)','W/F':'W/F Rating','SM':'SM Rating','IR':'IR Rating'},inplace=True)
```
